```python
df.drop(columns=['Student_ID'],inplace=True)
```

Outlier detection and treatment

```python
for col in df:
  if df[col].dtype=='int' or df[col].dtype=='float':
    print(col.capitalize())
    print("---------")
    print(f"Mean:{df[col].mean()}")
    print(f"Median:{df[col].median()}")
    print(f'Mode :',end='')
    for i in df[col].mode().to_list():
      print(i)
    print(f"Minimum of {col}:{df[col].min()}")
    print(f"Maximum of {col}:{df[col].max()}")
    print(f"Variance of {col}:{df[col].var()}")
    print(f"Standard deviation:{df[col].std()}")
    plt.figure(figsize=(8,5))
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
    print("___________________________________________________________________________________________________________________________________")
    print()
```

FEATURE ENGINEERING

```python
df['Age_Group'] = df['Age'].apply(lambda x: "Under 15" if x < 15 else ("15-18" if x <= 18 else ("18-21" if x <= 21 else "21+")))
```

```python
df['Age_Group']
```

```python
df['VR_Usage_Category'] = df['Hours_of_VR_Usage_Per_Week'].apply(lambda x: "Low Usage" if x <= 3 else ("Moderate Usage" if x <= 7 else ("High Usage" if x <= 15 else "Very High Usage")))
```

```python
df['Engagement_Effectiveness'] = df['Engagement_Level'] * df['Perceived_Effectiveness_of_VR']
```

```python
df['Total_VR_Exposure'] = df['Hours_of_VR_Usage_Per_Week'] * df['Engagement_Level']
```

UNIVARIATE ANALYSIS

```python
for col in df:
    if df[col].dtype == 'object':
        plt.figure(figsize=(8, 4))
        plt.bar(x=df[col].value_counts().index, height=df[col].value_counts())
        plt.title(col)
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()
        print(df[col].value_counts())
        print("___________________________________________________________________________________________________________________________________________________________________")
```

```python
for col in df:
    if df[col].dtype == 'object':
        plt.figure(figsize=(8,4))
        df[col].value_counts().plot.pie(autopct='%1.1f%%')
        plt.title(f'Distribution of {col}')
        plt.ylabel('')
        plt.show()
        print(df[col].value_counts())
        print("__________________________________________________________________________________________________________________________________________________________________")
```

```python
for col in df:
  if df[col].dtype == 'int' or df[col].dtype== 'float':
    print(col.capitalize())
    print("---------")
    print(f"Mean:{df[col].mean()}")
    print(f"Median:{df[col].median()}")
    print(f'Mode :',end='')
    for i in df[col].mode().to_list():
      print(i)
    print(f"Minimum of {col}:{df[col].min()}")
    print(f"Maximum of {col}:{df[col].max()}")
    print(f"Variance of {col}:{df[col].var()}")
    print(f"Standard deviation:{df[col].std()}")
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], bins=15)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

BIVARIATE ANALYSIS

```python
for i in df:
  if df[i].dtype=='object':
    for j in df:
      if df[j].dtype=='int' or df[j].dtype=='float':
        print(df[[i,j]].groupby(i).agg(['mean','median','max','min','std']))
        print('______________________________________________________________________________________________________________________________________________________________________')
        print()
```

```python
df.corr(numeric_only=True)
plt.figure(figsize=(8,4))
plt.title("Correlation Heatmap of Numerical Features")
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

*Student Distribution Evenness Across Grade Levels*

```python
grade_counts = df.groupby('Grade_Level').size()
std_dev = grade_counts.std() / grade_counts.mean()

print("Student Distribution Across Grade Levels:\n", grade_counts)
print("Standard Deviation (indicating evenness):", std_dev)
plt.figure(figsize=(8, 5))
grade_counts.plot(kind='bar', color='skyblue')
plt.title("Student Distribution Across Grade Levels")
plt.xlabel("Grade Level")
plt.ylabel("Number of Students")
plt.show()
```
